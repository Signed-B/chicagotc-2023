```python
# read and graph futures_2022.csv

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv('futures_2022.csv')
df.head()

# plot

plt.figure(figsize=(20,10))
plt.plot(df['date'], df['LBSA'], label='LBSA')
plt.plot(df['date'], df['LBSB'], label='LBSB')
plt.plot(df['date'], df['LBSC'], label='LBSC')
plt.plot(df['date'], df['LBSD'], label='LBSD')
plt.plot(df['date'], df['LBSE'], label='LBSE')
plt.plot(df['date'], df['LBSF'], label='LBSF')
plt.plot(df['date'], df['LBSG'], label='LBSG')
plt.plot(df['date'], df['LBSH'], label='LBSH')
plt.plot(df['date'], df['LBSI'], label='LBSI')
plt.plot(df['date'], df['LBSJ'], label='LBSJ')
plt.plot(df['date'], df['LBSK'], label='LBSK')
plt.plot(df['date'], df['LBSL'], label='LBSL')
plt.plot(df['date'], df['LBSM'], label='LBSM')
plt.plot(df['date'], df['LBSN'], label='LBSN')
plt.plot(df['date'], df['SBL'], label='SBL', color='black', linewidth=3)
plt.legend()
```
